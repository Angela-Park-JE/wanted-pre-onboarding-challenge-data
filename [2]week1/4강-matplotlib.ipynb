# 그래프를 그리는 2가지 방식
---
## 1. plt에 그리기
add_subplot()을 통해 다음 ax로 넘어가 그래프를 그립니다.

````python
fig = plt.figure(figsize=(6, 3)) # fig는 그래프 인스턴스들을 그릴 지면이고,

# 값이 x,y 값이 아니라서 알아서 y값으로 인식
x1 = [3, 1, 8]
x2 = [2, 7, 4]

# ax1에 그리기
ax1 = fig.add_subplot(1, 2, 1) # ax1는 그래프 인스턴스 하나를 의미 (row, column, first one)
plt.plot(x1) 

# ax2에 그리기
ax2 = fig.add_subplot(1, 2, 2) # (row, column, second one)
plt.plot(x2) 
plt.show()
````

## 2. ax에 직접 그리기
figure에 있는 각 ax에 직접적으로 그래프를 그립니다.

````python
# axes에 각 그래프 인스턴스가 들어있음(로우 컬럼 지정해줌)
fig, axes = plt.subplots(1,2, figsize=(6,3)) # 1행 2열 두 칸에 그릴 것으로 세팅

y1 = [3, 1, 8]
y2 = [2, 7, 4]

# axes의 인덱스에 접근하여 그리도록 한다
axes[0].plot(y1) # 0번: 첫번째
axes[1].plot(y2) # 1번: 두번째
print(f"axes type : {type(axes)}")
print(axes)
````

# 막대 그래프 (bar graph)
---

- 양을 비교할 때 사용
- 시각 속성 중 높이와 위치를 활용하며  
- 높이가 높을수록 값이 크고 
- 모든 막대가 0이라는 위치에서 시작함 (다르게 지정 가능하지만 왜곡 발생할 수 있으니 주의)

## 세로 막대그래프 : `ax.bar(x, y)`

````python
fig, ax = plt.subplots(1, 1, figsize=(4, 4)) # 그래프 한 개만 들어와있는 상태

# 데이터
x = [1, 2, 3, 4, 5]
y = np.array([9, 12, 8, 16, 13])

# 세로 막대 그래프
ax.bar(x, y)
````

### `bottom`옵션: 막대 그래프 "위치"의 중요성

(당연한 이야기이지만) 위치가 다른 곳에 있다면 비교하기 어려워진다. 항상 0에서부터 시작해야 한다. 

````python
fig, ax = plt.subplots(figsize=(5,5))

ax.bar(x, y, 
       bottom = np.random.choice(np.arange(5), size=5) # 바텀 5개에, 각각의 시작점을 랜덤하도록 어레이를 주었다.
        )
````

### 크기비교의 왜곡: 막대그래프가 0부터 시작해야하는 이유

5부터 시작한 그래프이다. 크기비교가 왜곡된다.

````python
fig, ax = plt.subplots(figsize=(4, 4))

ax.bar(x, y)
ax.set_ylim([5, 17])
# 8과 16이 2배 차이가 아니라 3배 차이나는 것으로 보이게 된다.
````

### color 옵션, label 옵션, 테두리 지우기, tick 지우기

````python
fig, ax = plt.subplots(figsize=(6, 6))

ax.bar(x, y, 
        color='gray', # 바 색 (바가 5개라서 5개 정해줄 수도 있음)
        label='label' # 범례
)

ax.set_title('Title') # 제목
ax.legend() # 범례를 할거면 쓰기
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False) # 그래프 테두리를 각각 다 지웠다.

ax.tick_params(left = False, bottom = False) # major tick들이 거슬릴 때 지우려면 이거 쓰기
# ax.set_yticks([]) # 숫자들도 지워버리고 싶으면 각 틱을 [] 빈 리스트로 줘버리면 된다.
# ax.set_xticks([])
ax.grid() # 틱이 없으면 그리드도 사라짐
````

## 예시 데이터 불러오기

특정 프로그래밍 언어를 사용하는 개발자의 연봉

Age : 나이 

All_Devs : 총 개발자 연봉

Python : Python을 쓰는 개발자의 연봉

JavaScript : JavaScript를 쓰는 개발자의 연봉

````python
import pandas as pd

df = pd.read_csv('age_data.csv')
df.head()
````

### 20대 개발자 전체를 세로 막대 그래프로 시각화해보세요.

````python
filtering_df = df[(df['Age']>=20) & (df['Age']<30)]

ages_x = filtering_df['Age']
dev_y = filtering_df['All_Devs']
py_dev_y = filtering_df['Python']

fig, ax = plt.subplots(figsize=(5, 4))

x_indexes = np.arange(len(ages_x))  # age로 index 만들음. 그냥 ages_x를 쓰기 되면 원하는 tick이 사라질 수 있다. 그러기에 억지로 넣어주는 것

ax.bar(x_indexes, dev_y, 
        color='gray', 
        label='All Devs')

# without this line our age range is lost, to fix this: 
ax.set_xticks(ticks=x_indexes, labels=ages_x) # 여기 써둔 틱들은 모두 나와야 한다. lables를 지우면 0~9로 써짐.
ax.legend()
````

### 파이썬 개발자와 전체 개발자의 나이대별 연봉을 비교해보세요

### Overlapped Bar Plot

- 투명도가 없을 경우, 더 작은 막대 그래프의 값은 볼 수 없습니다.

- 투명도를 조절하여 2개의 그래프를 비교해봅니다.  
겹치는 부분이 Legend에 없는 색깔이므로 헷갈릴 수 있는 여지가 있습니다.

선호하진 않는다. 겹치는 영역 때문에 보기 좋지도 않고, 더 나은게 그룹드 바 플롯.

````python
# side by side
fig, ax = plt.subplots(figsize=(6,5))
x_indexes = np.arange(len(ages_x)) 
bar_width = 0.35


ax.bar(x_indexes, 
       dev_y, # 데브올
       width=bar_width, # 두께
       color='#444444', # 색 지정
       alpha=0.5, # 투명도
       label='All Devs') 

ax.bar(x_indexes, 
       py_dev_y, # 파이썬들
       width=bar_width, 
       color='#008df5',
       alpha=0.5,
       label='Python')

# without this line our age range is lost, to fix this: 
ax.set_xticks(ticks=x_indexes, labels=ages_x)
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
ax.set_title("Median Salary (USD) by Age")

ax.legend()
# plt.tight_layout()
````

## Grouped Bar Plot

- Legend에 있는 색깔 2개로만 표현(색의 수 줄이기)
- Overllaped Bar Plot보다 인지하기 쉬움

틱을 기준으로 균등하게 그려지는데, 살짝 왼쪽 살짝 오른쪽에 오도록 그림.

````python
# side by side
fig, ax = plt.subplots(figsize=(6,5))
x_indexes = np.arange(len(ages_x)) 
bar_width = 0.35


ax.bar(x_indexes - bar_width/2,  # 두께의 절반만큼 앞으로 옮겨서
       dev_y, 
       width=bar_width, 
       color='#444444', 
       label='All Devs')

ax.bar(x_indexes + bar_width/2,  # 두께의 절반만큼 뒤로 옮겨서
       py_dev_y, 
       width=bar_width, 
       color='#008df5', 
       label='Python')

# without this line our age range is lost, to fix this: 
ax.set_xticks(ticks=x_indexes, labels=ages_x)
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
ax.set_title("Median Salary (USD) by Age")

ax.legend()
# plt.tight_layout()
````

## Percentage Bar Plot

- 전체 비율을 비교할 때 사용합니다.

공중부양 그래프 그렸던 것처럼 쌓는 식

````python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the MBTI data
data = pd.DataFrame({
    'PJ': [0.4, 0.6],
    'FT': [0.7, 0.3],
    'NS': [0.5, 0.5],
    'EI': [0.6, 0.4]
})

# Create a horizontal bar plot
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the ratio bar plot for 'ei'
ax.barh(data.columns, data.loc[0], height=0.5, color='blue') # bar'h' 호라이즌탈. 두께를 대신 'height'
ax.barh(data.columns, data.loc[1], height=0.5, left = data.loc[0], color='orange') # left 값을 정해서 시작점 지정

# Add labels to the plot
ax.set_xlabel('Percentage')
ax.set_ylabel('MBTI Types')
ax.set_title("MBTI")

for location in ['top','bottom','left','right']:
    ax.spines[location].set_visible(False) # spine 제거

# Show the plot
plt.show()
````

## `ax.barh(x, y)` : 가로 막대 그래프

(위에서 했지만 좀더 설명하자면!)  
가로 막대 그래프는 **비교해야할 대상이 많아질 때** 지면이 부족해 표현하기 힘든 점을 보완하여 독자의 인지를 더 쉽게 도와줍니다. 

````python
from collections import Counter # 자동으로 특정 글자나 그런게 있을 때 세서 딕셔너리 형태로 만들어준다.
df = pd.read_csv('data.csv') # 각 사람이 어떤 언어를 쓰는지에 대한 데이터
ids = df['Responder_id']
df.head()
````

````python
lang_responses = df['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';')) # Counter.update에 넣어주면 언어를 ;을 기준으로 리스트로 분리해줌. 언어가 몇 번 나왔는지 세어줌.
    

languages = []
popularity = []
for lan, pop in language_counter.most_common(15): # 가장 많이 나온 언어 15개 선정
    languages.append(lan) # 언어
    popularity.append(pop) # 언어별 몇명
````

````python
# plotting
fig, ax = plt.subplots()
languages.reverse() # 가로막대로 바꾸면 오름차순으로 그려지기 때문에 강제로 내림차순으로 정렬
popularity.reverse()
ax.barh(languages, popularity)
# plt.ylabel("Programming Languages")
ax.set_xlabel("Number of people who use")
ax.set_title("Most Popular Languages")

plt.tight_layout()
````

````python
from scipy.interpolate import make_interp_spline # 보간법 함수
import matplotlib.dates as dates # 맷플랏립 함수. 날짜를 숫자로 바꾸기 위해서

fig, ax = plt.subplots(1, 2, figsize=(20, 7), dpi=300)

date_np = google.index
value_np = google['close']

date_num = dates.date2num(date_np) # matplotlib 내장 함수. 날짜가 숫자로 되어 그릴 수 있도록 spline 함수를 만들기 위해 dates -> num 으로 변환

# smooth
date_num_smooth = np.linspace(date_num.min(), date_num.max(), 50) # 50개의 포인트를 가지고 그리도록. spline 함수 적용할 x 값 선정
spl = make_interp_spline(date_num, value_np) # 마감가 기준으로 spline을 그려줌 
value_np_smooth = spl(date_num_smooth)


# print
ax[0].plot(date_np, value_np) # 원래 날짜와 값
ax[1].plot(dates.num2date(date_num_smooth), value_np_smooth) # 숫자를 다시 날짜로 바꾸고 바뀐 값들로 그리도록


plt.show()
````

포인트를 찍기 위해선 스캐터플롯을 사용한다.

````python
fig = plt.figure(figsize=(7, 7))

np.random.seed(97)

x = np.arange(20)
y = np.random.rand(20)

fig, ax = plt.subplots()
ax.plot(x, y,
       color='lightgray',
       linewidth=2,)

ax.set_xlim(0, 21)

# max
# 회색 그래프를 그리기 위해 사용한 것 중에서 값을 가져와서 그리도록 한다.
ax.plot([0, x[np.argmax(y)]], [np.max(y)]*2, # y값이 최대일때의 x값을(인덱스를) 가져오고 x[np.argmax(y)]], y값 최대를 가져온다
        linestyle='--', color='tomato'
       )

ax.scatter(x[np.argmax(y)], np.max(y), 
            c='tomato',s=50, zorder=20)

# min
ax.plot([0, x[np.argmin(y)]], [np.min(y)]*2,
        linestyle='--', color='royalblue'
       )
ax.scatter(x[np.argmin(y)], np.min(y), 
            c='royalblue',s=50, zorder=20)

# 축에 써진 값
ax.text(x=-1.5, y=np.max(y), s=round(np.max(y),2),
        fontsize=10,
        fontweight='bold',
        fontfamily='serif',
        color = 'tomato',
        va='top' # vertical aline : 해당 포인트를 기준으로 top이도록
       )

ax.text(x=-1.5, y=np.min(y), s=round(np.min(y),2),
        fontsize=10,
        fontweight='bold',
        fontfamily='serif',
        color = 'royalblue',
        va='center'
       )

plt.show()
````

## 예시 데이터: 국가별 평균수명과 1인당 GDP의 관계

하이트로슬링? 팩트풀니스라는 책을 만든 분인데, 세상을 데이터로 바라보자고 하는 유명한 교수님이다.  
많은 업적을 남기셨는데, 그분의 유명한 데이터셋이다.

- Country : 국가
- Year : 연도
- Life Expectancy : 기대 수명
- Children per Woman : 출산율
- Population : 인구
- Continent : 대륙
- Income per Person : 소득
- Child Mortality : 아동 사망률
- 출처 : https://drive.google.com/file/d/1bm2ImU_MD-fvoNbIr-eTbAEUcipnfJZP/view

````python
data = pd.read_csv('hans_rosling_v4.csv')
data = data[data["Year"]=='Y2015']
data.head()
````

### 국가별로 데이터 살펴보기

````python
data['Country'].unique()
````

### 무의미한 색 분류 주의하기
시각 속성 중 색은 데이터 시각화에서 강력합니다. 하지만 잘 사용하지 못한다면 그 효과를 볼 수 없을 뿐더러 사람들에게 반감을 불러일으킬 수 있습니다.

````python
# Define colors for each country
country = data['Country'].unique()
colors = sns.color_palette("Set2", n_colors=len(country))

country_colors = {cou : color for cou, color in zip(country, colors)} # 국가와 색상별로 매칭

fig, axes = plt.subplots(1,2,figsize=(20,8))
# Create the scatter plot
axes[0].scatter(data['Life Expectancy'], data['Income per Person'],
            s=data['Population'], c=data['Country'].map(country_colors), label=data['Country']) # 사이즈에 인구수를 넣는것. 


# Add labels and title
axes[0].set_xlabel('기대 수명')
axes[0].set_ylabel('1인 소득')
axes[0].set_title('기대 수명 vs. 1인 소득')

# Add a legend for the country - 범례를 따로 그렸습니다.
# legend_handles -> 범례의 옵션만 따로 적어둔것.
legend_handles = [plt.Line2D([], [], marker='o', color='w', markersize=10, markerfacecolor=color, label=cou)
                  for cou, color in country_colors.items()]
# 두번째 그래프에 따로 범례만 표현
axes[1].legend(handles=legend_handles, title='Continent',loc='upper left')
for loc in ['top','bottom','left','right']:
    axes[1].spines[loc].set_visible(False)
axes[1].tick_params(left = False, bottom = False)
axes[1].set_xticks([])
axes[1].set_yticks([])
fig.tight_layout()
````

기대수명과 1인소득이 어느 정도 비례관계에 있다는 것은 보이지만...

모든 국가를 표현하는데 필요한 색의 수가 부족하고 범례가 길어 어떤 국가가 어떤 데이터를 의미하는지 파악하기 힘듭니다. 

### 색 한가지만 사용하기
위 그래프에서 어떤 정보를 파악할 수 있었나요?

- 경향성 파악 : 1인 소득이 높을수록 평균 수명이 높아집니다.

````python
# Create the scatter plot
fit, ax = plt.subplots()
ax.scatter(data['Life Expectancy'], data['Income per Person'],
            s=data['Population'], c='gray') # 상당히 간단함. 범례가 필요가 없는 그래프.

# Add labels and title
ax.set_xlabel('기대 수명')
ax.set_ylabel('1인 소득')
ax.set_title('기대 수명 vs. 1인 소득')
````

위 그래프를 통해 국가별로 색상을 나타낸 그래프와 달리 범례가 필요없고 경향성 정보를 파악하는데 충분합니다.

### 색깔 효율적으로 사용하기 -> 그룹화

대륙별로 구별하기 -> 색상에 각 나라의 위치, 공간 정보를 함축

````python
data['Continent'].unique()
````

````python
fig, ax = plt.subplots()

# Define colors for each continent
continent = data['Continent'].unique()
colors = sns.color_palette("Set2",n_colors=len(continent))

continent_colors = {cont : color for cont, color in zip(continent, colors)}

# Create the scatter plot
ax.scatter(data['Life Expectancy'], data['Income per Person'],
            s=data['Population'], c=data['Continent'].map(continent_colors)) # 대륙별로 색상을 표현하도록 했다.

# Add labels and title
ax.set_xlabel('기대 수명')
ax.set_ylabel('1인 소득')
ax.set_title('기대 수명 vs. 1인 소득')

# Add a legend for the continents
legend_handles = [plt.Line2D([], [], marker='o', color='w', markersize=10, markerfacecolor=color, label=continent)
                  for continent, color in continent_colors.items()]
ax.legend(handles=legend_handles, title='Continent')

````
