## 1. pltÏóê Í∑∏Î¶¨Í∏∞
add_subplot()ÏùÑ ÌÜµÌï¥ Îã§Ïùå axÎ°ú ÎÑòÏñ¥Í∞Ä Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶ΩÎãàÎã§.

````python
fig = plt.figure(figsize=(6, 3)) # figÎäî Í∑∏ÎûòÌîÑ Ïù∏Ïä§ÌÑ¥Ïä§Îì§ÏùÑ Í∑∏Î¶¥ ÏßÄÎ©¥Ïù¥Í≥†,

# Í∞íÏù¥ x,y Í∞íÏù¥ ÏïÑÎãàÎùºÏÑú ÏïåÏïÑÏÑú yÍ∞íÏúºÎ°ú Ïù∏Ïãù
x1 = [3, 1, 8]
x2 = [2, 7, 4]

# ax1Ïóê Í∑∏Î¶¨Í∏∞
ax1 = fig.add_subplot(1, 2, 1) # ax1Îäî Í∑∏ÎûòÌîÑ Ïù∏Ïä§ÌÑ¥Ïä§ ÌïòÎÇòÎ•º ÏùòÎØ∏ (row, column, first one)
plt.plot(x1) 

# ax2Ïóê Í∑∏Î¶¨Í∏∞
ax2 = fig.add_subplot(1, 2, 2) # (row, column, second one)
plt.plot(x2) 
plt.show()
````

## 2. axÏóê ÏßÅÏ†ë Í∑∏Î¶¨Í∏∞
figureÏóê ÏûàÎäî Í∞Å axÏóê ÏßÅÏ†ëÏ†ÅÏúºÎ°ú Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶ΩÎãàÎã§.

````python
# axesÏóê Í∞Å Í∑∏ÎûòÌîÑ Ïù∏Ïä§ÌÑ¥Ïä§Í∞Ä Îì§Ïñ¥ÏûàÏùå(Î°úÏö∞ Ïª¨Îüº ÏßÄÏ†ïÌï¥Ï§å)
fig, axes = plt.subplots(1,2, figsize=(6,3)) # 1Ìñâ 2Ïó¥ Îëê Ïπ∏Ïóê Í∑∏Î¶¥ Í≤ÉÏúºÎ°ú ÏÑ∏ÌåÖ

y1 = [3, 1, 8]
y2 = [2, 7, 4]

# axesÏùò Ïù∏Îç±Ïä§Ïóê Ï†ëÍ∑ºÌïòÏó¨ Í∑∏Î¶¨ÎèÑÎ°ù ÌïúÎã§
axes[0].plot(y1) # 0Î≤à: Ï≤´Î≤àÏß∏
axes[1].plot(y2) # 1Î≤à: ÎëêÎ≤àÏß∏
print(f"axes type : {type(axes)}")
print(axes)
````

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ (bar graph)

## ÏÑ∏Î°ú ÎßâÎåÄÍ∑∏ÎûòÌîÑ : `ax.bar(x, y)`

````python
fig, ax = plt.subplots(1, 1, figsize=(4, 4)) # Í∑∏ÎûòÌîÑ Ìïú Í∞úÎßå Îì§Ïñ¥ÏôÄÏûàÎäî ÏÉÅÌÉú

# Îç∞Ïù¥ÌÑ∞
x = [1, 2, 3, 4, 5]
y = np.array([9, 12, 8, 16, 13])

# ÏÑ∏Î°ú ÎßâÎåÄ Í∑∏ÎûòÌîÑ
ax.bar(x, y)
````

### `bottom`ÏòµÏÖò: ÎßâÎåÄ Í∑∏ÎûòÌîÑ "ÏúÑÏπò"Ïùò Ï§ëÏöîÏÑ±

(ÎãπÏó∞Ìïú Ïù¥ÏïºÍ∏∞Ïù¥ÏßÄÎßå) ÏúÑÏπòÍ∞Ä Îã§Î•∏ Í≥≥Ïóê ÏûàÎã§Î©¥ ÎπÑÍµêÌïòÍ∏∞ Ïñ¥Î†§ÏõåÏßÑÎã§. Ìï≠ÏÉÅ 0ÏóêÏÑúÎ∂ÄÌÑ∞ ÏãúÏûëÌï¥Ïïº ÌïúÎã§. 

````python
fig, ax = plt.subplots(figsize=(5,5))

ax.bar(x, y, 
       bottom = np.random.choice(np.arange(5), size=5) # Î∞îÌÖÄ 5Í∞úÏóê, Í∞ÅÍ∞ÅÏùò ÏãúÏûëÏ†êÏùÑ ÎûúÎç§ÌïòÎèÑÎ°ù Ïñ¥Î†àÏù¥Î•º Ï£ºÏóàÎã§.
        )
````

### ÌÅ¨Í∏∞ÎπÑÍµêÏùò ÏôúÍ≥°: ÎßâÎåÄÍ∑∏ÎûòÌîÑÍ∞Ä 0Î∂ÄÌÑ∞ ÏãúÏûëÌï¥ÏïºÌïòÎäî Ïù¥Ïú†

## `ax.barh(x, y)` : Í∞ÄÎ°ú ÎßâÎåÄ Í∑∏ÎûòÌîÑ

(ÏúÑÏóêÏÑú ÌñàÏßÄÎßå Ï¢ÄÎçî ÏÑ§Î™ÖÌïòÏûêÎ©¥!)  
Í∞ÄÎ°ú ÎßâÎåÄ Í∑∏ÎûòÌîÑÎäî **ÎπÑÍµêÌï¥ÏïºÌï† ÎåÄÏÉÅÏù¥ ÎßéÏïÑÏßà Îïå** ÏßÄÎ©¥Ïù¥ Î∂ÄÏ°±Ìï¥ ÌëúÌòÑÌïòÍ∏∞ ÌûòÎì† Ï†êÏùÑ Î≥¥ÏôÑÌïòÏó¨ ÎèÖÏûêÏùò Ïù∏ÏßÄÎ•º Îçî ÏâΩÍ≤å ÎèÑÏôÄÏ§çÎãàÎã§. 

````python
from collections import Counter # ÏûêÎèôÏúºÎ°ú ÌäπÏ†ï Í∏ÄÏûêÎÇò Í∑∏Îü∞Í≤å ÏûàÏùÑ Îïå ÏÑ∏ÏÑú ÎîïÏÖîÎÑàÎ¶¨ ÌòïÌÉúÎ°ú ÎßåÎì§Ïñ¥Ï§ÄÎã§.
df = pd.read_csv('data.csv') # Í∞Å ÏÇ¨ÎûåÏù¥ Ïñ¥Îñ§ Ïñ∏Ïñ¥Î•º Ïì∞ÎäîÏßÄÏóê ÎåÄÌïú Îç∞Ïù¥ÌÑ∞
ids = df['Responder_id']
df.head()
````

````python
lang_responses = df['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';')) # Counter.updateÏóê ÎÑ£Ïñ¥Ï£ºÎ©¥ Ïñ∏Ïñ¥Î•º ;ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Î¶¨Ïä§Ìä∏Î°ú Î∂ÑÎ¶¨Ìï¥Ï§å. Ïñ∏Ïñ¥Í∞Ä Î™á Î≤à ÎÇòÏôîÎäîÏßÄ ÏÑ∏Ïñ¥Ï§å.
    

languages = []
popularity = []
for lan, pop in language_counter.most_common(15): # Í∞ÄÏû• ÎßéÏù¥ ÎÇòÏò® Ïñ∏Ïñ¥ 15Í∞ú ÏÑ†Ï†ï
    languages.append(lan) # Ïñ∏Ïñ¥
    popularity.append(pop) # Ïñ∏Ïñ¥Î≥Ñ Î™áÎ™Ö
````

````python
# plotting
fig, ax = plt.subplots()
languages.reverse() # Í∞ÄÎ°úÎßâÎåÄÎ°ú Î∞îÍæ∏Î©¥ Ïò§Î¶ÑÏ∞®ÏàúÏúºÎ°ú Í∑∏Î†§ÏßÄÍ∏∞ ÎïåÎ¨∏Ïóê Í∞ïÏ†úÎ°ú ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨
popularity.reverse()
ax.barh(languages, popularity)
# plt.ylabel("Programming Languages")
ax.set_xlabel("Number of people who use")
ax.set_title("Most Popular Languages")

plt.tight_layout()
````

````python
from scipy.interpolate import make_interp_spline # Î≥¥Í∞ÑÎ≤ï Ìï®Ïàò
import matplotlib.dates as dates # Îß∑ÌîåÎûèÎ¶Ω Ìï®Ïàò. ÎÇ†ÏßúÎ•º Ïà´ÏûêÎ°ú Î∞îÍæ∏Í∏∞ ÏúÑÌï¥ÏÑú

fig, ax = plt.subplots(1, 2, figsize=(20, 7), dpi=300)

date_np = google.index
value_np = google['close']

date_num = dates.date2num(date_np) # matplotlib ÎÇ¥Ïû• Ìï®Ïàò. ÎÇ†ÏßúÍ∞Ä Ïà´ÏûêÎ°ú ÎêòÏñ¥ Í∑∏Î¶¥ Ïàò ÏûàÎèÑÎ°ù spline Ìï®ÏàòÎ•º ÎßåÎì§Í∏∞ ÏúÑÌï¥ dates -> num ÏúºÎ°ú Î≥ÄÌôò

# smooth
date_num_smooth = np.linspace(date_num.min(), date_num.max(), 50) # 50Í∞úÏùò Ìè¨Ïù∏Ìä∏Î•º Í∞ÄÏßÄÍ≥† Í∑∏Î¶¨ÎèÑÎ°ù. spline Ìï®Ïàò Ï†ÅÏö©Ìï† x Í∞í ÏÑ†Ï†ï
spl = make_interp_spline(date_num, value_np) # ÎßàÍ∞êÍ∞Ä Í∏∞Ï§ÄÏúºÎ°ú splineÏùÑ Í∑∏Î†§Ï§å 
value_np_smooth = spl(date_num_smooth)


# print
ax[0].plot(date_np, value_np) # ÏõêÎûò ÎÇ†ÏßúÏôÄ Í∞í
ax[1].plot(dates.num2date(date_num_smooth), value_np_smooth) # Ïà´ÏûêÎ•º Îã§Ïãú ÎÇ†ÏßúÎ°ú Î∞îÍæ∏Í≥† Î∞îÎÄê Í∞íÎì§Î°ú Í∑∏Î¶¨ÎèÑÎ°ù


plt.show()
````
