## Basic text commands

pyplot 혹은 ax 인스턴스로 그릴 수 있는 텍스트 API


|pyplot API|Object-oriented API|description|
|-|-|-|
|`text`|`text`|Axes text|
|`annoatate`|`annotate`|Axes annotation with arrow: 화살표 날리기|
|`suptitle`|`suptitle`|title of figure: 부제|
|`title`|`set_title`|title of subplot `ax`|
|`xlabel`|`set_xlabel`|x-axis label|
|`ylabel`|`set_ylabel`|y-axis label|
|`figtext`|`text`|figure text|

- 출처 : https://matplotlib.org/stable/tutorials/text/text_intro.html
- 함께 보면 좋은 자료 : https://m2.material.io/design/typography/understanding-typography.html#type-properties

```python
import matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.subplots_adjust(top=0.85)

# Set titles for the figure and the subplot respectively
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
ax.set_title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

# Set both x- and y-axis limits to [0, 10] instead of default [0, 1]
ax.axis([0, 10, 0, 10])

ax.text(3, 8, 'boxed italics text in data coords', style='italic', # 이탤릭체로 넣은 박스
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})

ax.text(2, 6, 'an equation: $E=mc^2$', fontsize=15)

ax.text(3, 2, 'Unicode: Institut für Festkörperphysik')

ax.text(9.5, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        color='green', fontsize=15)

ax.plot([2], [1], 'o') # 특정 포인트
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4), # 특정 포인트에 화살표!
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

## Labels for x- and y-axis

x축과 y축의 label을 지정할 수 있는 함수
- `~matplotlib.axes.Axes.set_xlabel` 
- `~matplotlib.axes.Axes.set_ylabel`




```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Damped oscillation [V]')

plt.show()
```

x, y_label은 x,y_ticklabels가 잘 보이도록 자동으로 배치되는데요. 아래와 위의 그래프를 비교해보세요.


```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1*10000)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Damped oscillation [V]')

plt.show()
```

만일, label과 axes를 떨어뜨리고자 한다면 lablepad를 줍니다.

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1*10000)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Damped oscillation [V]', labelpad=18) # 그래프와 약간의 떨어지도록

plt.show()
```

x_label의 위치를 변경하고 싶다면 `position` 키워드를 사용할 수 있습니다. 이 때, position의 인자 값은 x값만 조정 가능하고 y 값은 어떤 값이 들어오더라도 영향을 주지 않습니다. y 값을 변경하고자 한다면 `labelpad`를 사용하세요.


```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
ax.set_xlabel('Time [s]', position=(0., 1e6), horizontalalignment='left') # 포지션을 지정한 것
ax.set_ylabel('Damped oscillation [V]')

plt.show()
```

`matplotlib.font_manager.FontProperties`이나 `matplotlib.axes.Axes.set_xlabel`에 인자값을 넣어줌으로써 폰트를 바꿔줄 수 있습니다.

```python
from matplotlib.font_manager import FontProperties

font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_style('italic')

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
ax.set_xlabel('Time [s]', fontsize='large', fontweight='bold')
ax.set_ylabel('Damped oscillation [V]', fontproperties=font)

plt.show()
```

LaTeX를 통해 수학 기호를 삽입할 수 있습니다.

참고) https://matplotlib.org/stable/tutorials/text/usetex.html

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.2, left=0.2)
ax.plot(x1, np.cumsum(y1**2))
ax.set_xlabel('Time [s] \n This was a long experiment')
ax.set_ylabel('$\int\ Y^2\ dt\ \ [V^2 s]$')
plt.show()
```

## Titles

subplot의 title 또한 많은 경우 label과 같은 방식으로 설정할 수 있습니다만 *loc* 인자를 통해 위치를 조정할 수 있습니다. 기본 설정은 ``loc=center``입니다.

```python
fig, axs = plt.subplots(3, 1, figsize=(5, 6), tight_layout=True)
locs = ['center', 'left', 'right']
for ax, loc in zip(axs, locs):
    ax.plot(x1, y1)
    ax.set_title('Title with loc at '+loc, loc=loc)
plt.show()
```

title의 아래 공간은 pad를 통해 조절할 수 있습니다.
다른 값을 넣어보세요.

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(top=0.8)
ax.plot(x1, y1)
ax.set_title('Vertically offset title', pad=30) # 띄우기
plt.show()
```

## Ticks와 ticklabels

ticks와 ticklabels는 figure에서 다루기 어려운 측면이 있습니다. matplotlib에서는 이를 최선의 방식으로 자동 설정해줍니다. 하지만 사용자가 tick의 위치와 label하고자 하는 방식을 선택할 수 있도록 유연하게 사용할 수도 있습니다.

- `ax.xaxis`와 `ax.yaxis`는 축에 있는 label의 설정 정보를 포함하고 있습니다.
- `.Axis.set_major_locator`와 `.Axis.set_minor_locator`는 major와 minor ticks의 위치를 결정하는데 사용됩니다. 
- `.Axis.set_major_formatter` 와 `.Axis.set_minor_formatter`는 tick labels의 format을 결정합니다.

### Simple ticks

tick의 값을 자동으로 정의하는 기능은 편리하지만 이따금 기본 locators와 formatters가 덮어써질 때가 있습니다. 위의 그래프와 비교했을 때 이는 그래프의 가시성을 해친다는 것을 알 수 있습니다. 또한 axis의 구간을 임의로 변경할 경우 자동으로 구간이 설정된 첫번째 그래프와 다르게 2번째는 구간을 벗어난 것을 볼 수 있습니다.

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
axs[1].xaxis.set_ticks(np.arange(0., 8.1, 2.))
plt.show()
```

아래 예시 또한 tick의 format을 변경한 것입니다.


```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
ticks = np.arange(0., 8.1, 2.)
# list comprehension to get all tick labels...
tickla = [f'{tick:1.2f}' for tick in ticks] 
axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_ticklabels(tickla) # 원하는 숫자를 만들어서 직접 넣는 방식
axs[1].set_xlim(axs[0].get_xlim())
plt.show()
```

### Tick Locators and Formatters

틱의 위치 

ticklabels 리스트를 만드는 것 대신에 `matplotlib.ticker.StrMethodFormatter`를 사용하여 label의 string format을 설정할 수 있습니다. 

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
ticks = np.arange(0., 8.1, 2.)
axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_major_formatter('{x:1.1f}')
axs[1].set_xlim(axs[0].get_xlim())
plt.show()
```

`matplotlib.ticker.FixedLocator`로 눈금을 지정하고 `set_major_locator`를 사용하여 tick을 변경합니다. 위의 그래프에서 x축 구간을 가져오지 않더라도 자동으로 구간을 설정해줄 수 있습니다. 

`set_major_formatter`를 사용하여 string format을 변경합니다.


```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
locator = matplotlib.ticker.FixedLocator(ticks)
axs[1].xaxis.set_major_locator(locator)
axs[1].xaxis.set_major_formatter('±{x}°')
plt.show()
```

`matplotlib.ticker.StrMethodFormatter` 대신에 사용자 함수를 만들어 `ax.xaxis.set_major_formatter`에 사용할 수 있습니다.

```python
# 홀수일때 넣고 싶을 때 
def formatoddticks(x, pos):
    """Format odd tick positions."""
    if x % 2:
        return f'{x:1.2f}'
    else:
        return ''


fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
ax.plot(x1, y1)
ticks = np.arange(0,6,1)
locator = matplotlib.ticker.FixedLocator(ticks) # FixedLocator 지정한 구간 내 틱이 무조건 있도록 지정해주는 것
ax.xaxis.set_major_formatter(formatoddticks)
ax.xaxis.set_major_locator(locator)

plt.show()
```

### Dateticks

matplotlib은 그래프를 그릴 때 `datetime.datetime`와 `numpy.datetime64`를 인식하여 데이터로 사용할 수 있습니다. 각 tick이 겹치지 않게 rotation을 넣어줍니다.


```python
import datetime

fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
base = datetime.datetime(2023, 1, 1, 0, 0, 1)
time = [base + datetime.timedelta(days=x) for x in range(len(x1))]

ax.plot(time, y1)
ax.tick_params(axis='x', rotation=20)
```

`matplotlib.dates.DateFormatter`를 사용하여 어떤 format으로 날짜를 표현할 것인지 명시합니다. `.dates.DayLocator`를 사용하여 특정 날짜의 tick을 지정할 수 있습니다.

```python
import matplotlib.dates as mdates

locator = mdates.DayLocator(bymonthday=[1, 15])
formatter = mdates.DateFormatter('%b %d')

fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
ax.plot(time, y1)
ax.tick_params(axis='x')
plt.show()
```

## Text Layout

Tex를 특정한 위치를 주지 않고 간편하게 Layout을 사용하여 배치할 수 있습니다.
transform에 ax.transAxes를 주게 되면 왼쪽 아래는 (0,0), 오른쪽 위는 (1,1)로 축 변환이 됩니다.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# build a rectangle in axes coords
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# axes coordinates: (0, 0) is bottom left and (1, 1) is upper right
p = patches.Rectangle(
    (left, bottom), width, height,
    fill=False, transform=ax.transAxes
    )

ax.add_patch(p)

ax.text(left, bottom, 'left top',
        horizontalalignment='left',
        verticalalignment='top', # verticalalignment = va 둘중 아무거나 써도됨
        transform=ax.transAxes)

ax.set_axis_off()
plt.show()
```

아래를 실행하면 각 레이아웃 별로 어떻게 출력이 되는지 볼 수 있다!

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# build a rectangle in axes coords
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# axes coordinates: (0, 0) is bottom left and (1, 1) is upper right
p = patches.Rectangle(
    (left, bottom), width, height,
    fill=False, transform=ax.transAxes
    )

ax.add_patch(p)

ax.text(left, bottom, 'left top',
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes)

ax.text(left, bottom, 'left bottom',
        horizontalalignment='left',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(right, top, 'right bottom',
        horizontalalignment='right',
        verticalalignment='bottom',
        transform=ax.transAxes)

ax.text(right, top, 'right top',
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes)

ax.text(right, bottom, 'center top',
        horizontalalignment='center',
        verticalalignment='top',
        transform=ax.transAxes)

ax.text(left, 0.5*(bottom+top), 'right center',
        horizontalalignment='right',
        verticalalignment='center',
        rotation='vertical',
        transform=ax.transAxes)

ax.text(left, 0.5*(bottom+top), 'left center',
        horizontalalignment='left',
        verticalalignment='center',
        rotation='vertical',
        transform=ax.transAxes)

ax.text(0.5*(left+right), 0.5*(bottom+top), 'middle',
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=20, color='red',
        transform=ax.transAxes)

ax.text(right, 0.5*(bottom+top), 'centered',
        horizontalalignment='center',
        verticalalignment='center',
        rotation='vertical',
        transform=ax.transAxes)

ax.text(left, top, 'rotated\nwith newlines',
        horizontalalignment='center',
        verticalalignment='center',
        rotation=45,
        transform=ax.transAxes)

ax.set_axis_off()
plt.show()
```

## Basic annotation

주석은 2가지 고려해야 할 부분이 있습니다.
- 실제 데이터의 위치 (xy)
- 주석의 시작 위치 (xytext)

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(3, 3))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5), # xy 위치와 해당 xy에 넣을 실제 주석의 위치
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_ylim(-2, 2)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xycoords='data',
            xytext=(0.1, 0.99), textcoords='axes fraction',
            va='top', ha='left',
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.set_ylim(-2, 2)
```

## 색상 이해하기


- **Hue(색조)** : 빨강, 파랑, 초록 등 색상으로 생각하는 부분
    - 빨강에서 보라색까지 있는 스펙트럼에서 0-360으로 표현합니다.
- **Saturation(채도)** : 색의 선명도 
    - 색이 선명할수록 채도가 높다고 말하며 회색이나 흰색 또는 검정과 같은 무채색에 가까울수록 채도가 낮다고 말합니다(선명하다와 탁하다).
- **Lightness(광도)** : 색상의 밝기, V로도 얘기하여 컴퓨터에서 HSV 로 표현된다.

## Color formats

|description|example|
|-|-|
|RGB or RGBA (red, green, blue, alpha) tuple of float values in a closed interval.|`(0.1, 0.2, 0.5)`,`(0.1, 0.2, 0.5, 0.3)`|
|Case-insensitive hex RGB or RGBA string.: 16진법 RGB |`'#0f0f0f'`,`'#0f0f0f80'`|

아래 코드는 Matplotlib 라이브러리를 사용하여 색상 표를 그리는 함수입니다. 색상 목록(colors)을 기반으로 표를 생성하며, 선택적으로 열의 수(ncols)와 색상 정렬 여부(sort_colors)를 설정할 수 있습니다.

함수 내부에서는 각각의 칸(cell)의 크기, 여백(margin), 행과 열의 수(nrows, ncols) 등을 계산하고, 그림(figure)과 축(axes)을 생성합니다. 그 후에는 축을 표시하지 않고, 각 칸에 해당하는 색상과 이름을 텍스트로 표시하고, 해당 색상으로 직사각형 패치를 그립니다. 마지막으로 생성된 그림(figure)을 반환합니다.

이 코드는 색상 목록을 시각화하는 데 유용하며, 색상 표를 만들어 특정 색상을 시각적으로 비교하고 싶을 때 사용할 수 있습니다.

```python
import math

from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return 
```

```python
plot_colortable(mcolors.BASE_COLORS, ncols=3, sort_colors=False)
```

```python
plot_colortable(mcolors.TABLEAU_COLORS, ncols=2, sort_colors=False) # 태블로에서 사용되는 색상들. 상당히 유용합니당.
```

```python
plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

```python
plot_colortable(mcolors.XKCD_COLORS)
```

### Comparison between X11/CSS4 and xkcd colors

```python
import matplotlib.colors as mcolors
import matplotlib.patches as mpatch

overlap = {name for name in mcolors.CSS4_COLORS
           if f'xkcd:{name}' in mcolors.XKCD_COLORS}

fig = plt.figure(figsize=[9, 5])
ax = fig.add_axes([0, 0, 1, 1])

n_groups = 3
n_rows = len(overlap) // n_groups + 1

for j, color_name in enumerate(sorted(overlap)):
    css4 = mcolors.CSS4_COLORS[color_name]
    xkcd = mcolors.XKCD_COLORS[f'xkcd:{color_name}'].upper()

    # Pick text colour based on perceived luminance.
    rgba = mcolors.to_rgba_array([css4, xkcd])
    luma = 0.299 * rgba[:, 0] + 0.587 * rgba[:, 1] + 0.114 * rgba[:, 2]
    css4_text_color = 'k' if luma[0] > 0.5 else 'w'
    xkcd_text_color = 'k' if luma[1] > 0.5 else 'w'

    col_shift = (j // n_rows) * 3
    y_pos = j % n_rows
    text_args = dict(fontsize=10, weight='bold' if css4 == xkcd else None)
    ax.add_patch(mpatch.Rectangle((0 + col_shift, y_pos), 1, 1, color=css4))
    ax.add_patch(mpatch.Rectangle((1 + col_shift, y_pos), 1, 1, color=xkcd))
    ax.text(0.5 + col_shift, y_pos + .7, css4,
            color=css4_text_color, ha='center', **text_args)
    ax.text(1.5 + col_shift, y_pos + .7, xkcd,
            color=xkcd_text_color, ha='center', **text_args)
    ax.text(2 + col_shift, y_pos + .7, f'  {color_name}', **text_args)

for g in range(n_groups):
    ax.hlines(range(n_rows), 3*g, 3*g + 2.8, color='0.7', linewidth=1)
    ax.text(0.5 + 3*g, -0.3, 'X11/CSS4', ha='center')
    ax.text(1.5 + 3*g, -0.3, 'xkcd', ha='center')

ax.set_xlim(0, 3 * n_groups)
ax.set_ylim(n_rows, -1)
ax.axis('off')

plt.show()
```

### Basic continuous colorbar : 연속적 컬러바
`colorbar`는 norm과 cmap을 인자값으로 갖는 `ScalarMappable`을 호출합니다.

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

fig, ax = plt.subplots(figsize=(6, 1))
fig.subplots_adjust(bottom=0.5)

cmap = mpl.cm.cool # cmap의 cool color bar
norm = mpl.colors.Normalize(vmin=5, vmax=10) # vmin vmax값이 달라질 수 있기 때문에 Normalize 를 같이 써주어야 한다.

fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
             cax=ax, orientation='horizontal', label='Some Units')
```

### Extended colorbar with continuous colorscale : 이산적 컬러바
경계선을 기준으로 색깔을 이산적으로 변화를 줄 수 있으며, `extend` 인자로 양 끝 경계선 이후에도 색깔을 설정할 것인지 선택할 수 있습니다.{'neither', 'both', 'min', 'max'}

```python
fig, ax = plt.subplots(figsize=(6, 1))
fig.subplots_adjust(bottom=0.5)

cmap = mpl.cm.viridis
bounds = [-1, 2, 5, 7, 12, 15]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='both')

fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
             cax=ax, orientation='horizontal',
             label="Discrete intervals with extend='both' keyword")
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from colorspacious import cspace_converter
```

```python
cmaps = {}

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))


def plot_color_gradients(category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axs = plt.subplots(nrows=nrows + 1, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh,
                        left=0.2, right=0.99)
    axs[0].set_title(f'{category} colormaps', fontsize=14)

    for ax, name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=mpl.colormaps[name])
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()

    # Save colormap list for later.
    cmaps[category] = cmap_list
```

```python
plot_color_gradients('Perceptually Uniform Sequential',
                     ['viridis', 'plasma', 'inferno', 'magma', 'cividis'])
```
